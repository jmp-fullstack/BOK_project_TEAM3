```python
import requests
import json
import pandas as pd
from bs4 import BeautifulSoup
```

```python
URL = 'https://www.bok.or.kr/portal/singl/baseRate/list.do?dataSeCd=01&menuNo=200643'
response = requests.get(URL)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
datas = soup.select('td')
datas 
```

```python
# 연도 
for i in range(len(datas)):
    if i % 3 != 0:
        continue
    num = i * 3 
    if num > len(datas):
        break
    print(datas[3 * i].text+'-')
```

```python
# 월일
for i in range(len(datas)):
    if i % 3 != 0:
        continue
    num = i * 3 
    if num > len(datas):
        break
    print(datas[(3 * i)+1].text.replace('월', '-').replace('일', '').replace(' ', ''))
```

```python
# 기준 금리
for i in range(len(datas)):
    if i % 3 != 0:
        continue
    num = i * 3 
    if num > len(datas):
        break
    print(datas[(3 * i)+2].text)
```

```python
rates = []
for i in range(len(datas)):
    if i % 3 != 0:
        continue
    num = i * 3 
    if num > len(datas):
        break
    rates.append({
        '날짜': datas[3 * i].text+'-' + datas[(3 * i)+1].text.replace('월', '-').replace('일', '').replace(' ', ''),
        '기준금리': datas[(3 * i) + 2].text,

    })
rates

columns = ','.join(rates[0].keys())
# print(columns)
columns = ['날짜', '기준금리']
values = []
for rate in rates:
    values.append(tuple(rate.values()))
values

f = open('Basemoney_rate.csv', 'w', encoding='utf-8')
f.write('날짜\t기준금리\n')
for rate in rates:
    f.write( rate['날짜'] + '\t' + rate['기준금리'] + '\n' )

f.close()
```
