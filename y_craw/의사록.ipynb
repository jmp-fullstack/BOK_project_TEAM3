```python
import requests
import os
from bs4 import BeautifulSoup

# 다운로드할 파일을 저장할 폴더 경로
folder_path = 'Minutes_files'

download_count = 0
page_number = 0

# 1페이지부터 41페이지까지 pdf 데이터를 가져옴
for i in range(1, 41 + 1):
    URL = 'https://www.bok.or.kr/portal/bbs/B0000245/list.do?menuNo=200761'
    params = {
        'pageIndex': i
    }
    response = requests.get(URL, params=params)
    soup = BeautifulSoup(response.content, 'html.parser')

    # 현재 페이지의 li 요소들을 가져와서 처리
    li_list = soup.find('div', class_="bdLine type2").find('ul').find_all('li')

    for j in range(len(li_list)):
        try:
            li = li_list[j]
            file_group_box = li.find('div', class_='fileGoupBox')
            if file_group_box:
                link = file_group_box.find('ul').find('li').find('a').attrs['href']
                title = li.find('div', class_='row').find('span').find('a').find('span').find('span').text
                url2 = 'https://www.bok.or.kr' + link
                file_res = requests.get(url2)
                
                # 파일을 저장할 경로 설정
                file_path = os.path.join(folder_path, f"{title}.pdf")

                with open(file_path, 'wb') as f:
                    f.write(file_res.content) 

                download_count += 1

        except Exception as e:
            print(f"다음 요소를 처리하는 중 오류가 발생했습니다: {e}")

    # 10개의 파일을 다운로드할 때마다 페이지 숫자를 증가시킴
    if download_count % 10 == 0 and download_count != 0:
        page_number += 1
        print(f"{page_number}페이지입니다")

    print(f"현재 다운로드한 파일 개수: {download_count}")
```

```python
import os
import requests
import json
import pandas as pd
from bs4 import BeautifulSoup
import PyPDF2
```

```python
li_list = []

# 1페이지부터 41페이지까지 데이터를 가져옴
for i in range(1, 41 + 1):
    URL = 'https://www.bok.or.kr/portal/bbs/B0000245/list.do?menuNo=200761'
    params = {
        'pageIndex': i
    }
    response = requests.get(URL, params=params)
    soup = BeautifulSoup(response.content, 'html.parser')

    # 현재 페이지에서 li 요소들을 추출하여 li_list에 추가
    li = soup.find('div', class_="bdLine type2").find('ul').find_all('li')
    li_list.extend(li)

len(li_list)
```

```python
# 다운로드할 파일을 저장할 폴더 경로
folder_path = 'Minutes_file_1'

# 폴더가 존재하지 않으면 생성
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

# 페이지 알림
for j in range(len(li_list)):
    try:
        link = li_list[j].find('div', class_='fileGoupBox').find('ul').find('li').find('a').attrs['href']
        title = li_list[j].find('div', class_='row').find('span').find('a').find('span').find('span').text.strip()
        if 'javascript:void(0);' not in link:  # 올바른 링크인지 확인
            url2 = 'https://www.bok.or.kr' + link
            file_res = requests.get(url2)
            file_path = os.path.join(folder_path, f"{title}.pdf")
            with open(file_path, 'wb') as f:
                f.write(file_res.content)

            # PDF 파일 읽기
            with open(file_path, 'rb') as pdf_file:
                pdf_reader = PyPDF2.PdfFileReader(pdf_file)
                num_pages = pdf_reader.numPages
                for page_num in range(num_pages):
                    page = pdf_reader.getPage(page_num)
                    text = page.extractText()
                    print(f"Page {page_num + 1}:")
                    print(text)
        else:
            print(f"Invalid link found for '{title}': {link}")

    except Exception as e:
        print(f"Error occurred while processing '{title}': {e}")

print("Done")
```

```python
# pdf 추출
from PyPDF2 import PdfReader

# PDF 파일 경로
pdf_path = '금융통화위원회 의사록(2004년도 제1차)(2004.1.8).pdf'

# PDF 파일 열기
with open(pdf_path, 'rb') as file:
    # PdfReader를 사용하여 PDF 파일을 읽기
    pdf_reader = PdfReader(file)
    
    # 페이지 수 출력
    print(f"총 페이지 수: {len(pdf_reader.pages)}")

    # 각 페이지의 내용 출력
    for page_number, page in enumerate(pdf_reader.pages, start=1):
        print(f"페이지 {page_number} 내용:")
        print(page.extract_text())
```
